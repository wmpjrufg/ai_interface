```python
import pandas as pd
import pickle
from sklearn.metrics import r2_score

# 1. Carregar o dataset
df = pd.read_csv('drug200.csv') #[cite: 1]

# 2. Mapeamento da coluna 'Drug' para numérica (ajuste os valores conforme sua preferência)
mapeamento_drug = {
    'DrugY': 0,
    'drugC': 1,
    'drugX': 2,
    'drugA': 3,
    'drugB': 4
}
df['Drug_num'] = df['Drug'].map(mapeamento_drug)

# 3. Carregar o modelo (Certifique-se de que o modelo foi treinado com os mesmos inputs)
with open('drug200_Random_Forest.pkl', 'rb') as f:
    modelo = pickle.load(f)

print("Modelo carregado!")

# 4. Tratamento dos dados de entrada (X)
# Removemos as colunas de saída para preparar a matriz de entrada
X_input = df.drop(columns=['Drug', 'Drug_num'])
X_input = pd.get_dummies(X_input, drop_first=True)

# 5. Previsão
previsao = modelo.predict(X_input)
df['predict'] = previsao

# 6. Cálculo do R²
# Comparamos o alvo numérico criado com as previsões do modelo
r2 = r2_score(df['Drug_num'], df['predict'])
print(f"O R² do modelo neste dataset é: {r2:.4f}")

# 7. Exportar e exibir
df.to_excel('machine_learning_r02.xlsx', index=False)
df_comparacao = df[['Drug_num', 'predict']]
display(df_comparacao)
```
