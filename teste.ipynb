```python
import pandas as pd
import dill as pickle
from sklearn.metrics import r2_score

# Carregar o dataset
df = pd.read_csv('dados/drug200.csv') #[cite: 1]

# Mapeamento da coluna 'Drug' para numérica (ajuste os valores conforme sua preferência)
mapeamento_drug = {
    'DrugY': 0,
    'drugC': 1,
    'drugX': 2,
    'drugA': 3,
    'drugB': 4
}
df['Drug_num'] = df['Drug'].map(mapeamento_drug)

# Carregar o modelo (Certifique-se de que o modelo foi treinado com os mesmos inputs)
with open('dados/reg_gradient_boosting.pkl', 'rb') as f:
    modelo = pickle.load(f)

print("Modelo carregado!")

# Tratamento dos dados de entrada (X)
# Removemos as colunas de saída para preparar a matriz de entrada
X_input = df.drop(columns=['Drug', 'Drug_num'])
X_input = pd.get_dummies(X_input, drop_first=True)

# Previsão
previsao = modelo.predict(X_input)
df['predict'] = previsao

# Cálculo do R²
# Comparamos o alvo numérico criado com as previsões do modelo
r2 = r2_score(df['Drug_num'], df['predict'])
print(f"O R² do modelo neste dataset é: {r2:.4f}")

# Exportar e exibir
df.to_excel('machine_learning_r02.xlsx', index=False)
df_comparacao = df[['Drug_num', 'predict']]
display(df_comparacao)
```

```python
import pandas as pd
import dill as pickle
from sklearn.metrics import r2_score

# Carregar o dataset
df = pd.read_csv('dados/Car_Evaluation_Dataset.csv')

# Mapeamento da coluna alvo 'class' para numérica
mapeamento_class = {
    'unacc': 0,
    'acc': 1,
    'good': 2,
    'vgood': 3
}
df['class_num'] = df['class'].map(mapeamento_class)

# Carregar o modelo
# Exemplo: 'car_evaluation_Random_Forest.pkl'
nome_do_arquivo_pkl = 'dados/reg_random_forest.pkl' 
with open(nome_do_arquivo_pkl, 'rb') as f:
    modelo = pickle.load(f)

print("Modelo carregado!")

# Tratamento dos dados de entrada (X)
# Removemos as colunas de saída (texto e numérica) para preparar a matriz de entrada
X_input = df.drop(columns=['class', 'class_num'])
X_input = pd.get_dummies(X_input, drop_first=True)

# Previsão
previsao = modelo.predict(X_input)
df['predict'] = previsao

# Cálculo do R²
# Comparamos o alvo numérico criado com as previsões do modelo
r2 = r2_score(df['class_num'], df['predict'])
print(f"O R² do modelo neste dataset é: {r2:.4f}")

# Exportar e exibir
df.to_excel('machine_learning_car_evaluation.xlsx', index=False)
df_comparacao = df[['class_num', 'predict']]
display(df_comparacao)
```
